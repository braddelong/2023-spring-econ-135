```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
```

```python
g = []
n = []
epsilon = []
L = [1]
y = [1.25]

h = 0.0005
gamma = 2
beta = 0.025

for t in range(999):
    epsilon = epsilon + [rnd.normalvariate(0, 0.02)]
    n = n + [beta * (y[t] - 1) + epsilon[t]]
    g = g + [h - n[t]/gamma + rnd.normalvariate(0, 0.0005)]
    y = y + [y[t]*np.exp(g[t])]
    L = L + [L[t]*np.exp(n[t])]
```

```python
tseries = range(1000)
fig, ax = plt.subplots()

ax.plot(tseries, L, y)
```

```python
fig, ax = plt.subplots()

ax.plot(L, y)
```
